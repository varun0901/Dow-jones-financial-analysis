# Dow Jones stock analysis

```python
#get the information from the ticker
info2 = []

for i in ticker_symbol:
    info2.append(yf.Ticker(i).info)
```

```python
#convert information to DataFrame
df_dow = pd.DataFrame(info2)
df_dow
```

```python
# setting up symbol as index
df_dow.set_index('symbol')
```

```python
#printing tha Current Ratio
current_rario = df_dow["currentRatio"]
```

```python
#ploting this data into bar Chart
plt.figure(figsize=(16,5))
plt.bar(df_dow["shortName"], df_dow["currentRatio"])
plt.xticks(rotation=90)
plt.show()
```
